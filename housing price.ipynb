```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv("E:/kc_house_data.csv",index_col=0) #reading datas and indexing
df.head()
```

```python
price1=np.array(df['price'])
living_area=np.array(df['sqft_living'])    #plt.plot line plot
plt.scatter(price1,living_area)              #plt.scatter give scatter plot
plt.title('housing price')

plt.xlabel("income")
plt.ylabel('number')
```
